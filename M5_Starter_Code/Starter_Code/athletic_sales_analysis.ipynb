#### Combine the sales data by rows.

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
# Show the top 5 results.
products_sold = combined_df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()
products_sold.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)
top_5_sold = products_sold.sort_values(by='Total_Products_Sold', ascending=False).head(5)
top_5_sold
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city
# Rename the "units_sold" column to "Total_Products_Sold"
# Show the top 5 results.
products_sold_pivot = pd.pivot_table(combined_df, values='units_sold', index=['region', 'state', 'city'],aggfunc='sum').reset_index()
products_sold_pivot.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)
top_5_sold_pivot = products_sold_pivot.sort_values(by='Total_Products_Sold', ascending=False).head(5)
top_5_sold_pivot
```

```python
products_sold_pivot
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
# Show the top 5 results.
total_sales = combined_df.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()
total_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
top_5_sales = total_sales.sort_values(by='Total Sales', ascending=False).head(5)
top_5_sales
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
# Optional: Rename the "total_sales" column to "Total Sales"
# Show the top 5 results.
total_sales_pivot = pd.pivot_table(combined_df, values='total_sales', index=['region', 'state', 'city'],aggfunc='sum').reset_index()
total_sales_pivot.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
top_5_sales_pivot = total_sales_pivot.sort_values(by='Total Sales', ascending=False).head(5)
top_5_sales_pivot
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
# Show the top 5 results.
total_sales_retailer = combined_df.groupby(['retailer', 'region', 'state', 'city'])['total_sales'].sum().reset_index()
total_sales_retailer.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
top_5_sales_retailer = total_sales_retailer.sort_values(by='Total Sales', ascending=False).head(5)
top_5_sales_retailer
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Optional: Rename the "total_sales" column to "Total Sales"
# Show the top 5 results.
total_sales_retailer_pivot = pd.pivot_table(combined_df, values='total_sales', index=['retailer', 'region', 'state', 'city'],aggfunc='sum').reset_index()
total_sales_retailer_pivot.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
top_5_sales_retailer_pivot = total_sales_retailer_pivot.sort_values(by='Total Sales', ascending=False).head(5)
top_5_sales_retailer_pivot
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_products = combined_df[combined_df['product'].str.contains("Women", case=False, na=False)]
womens_products
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
# Show the top 5 results.
womens_products_sold = womens_products.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()
womens_products_sold.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)
top_5_womens_products_sold = womens_products_sold.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)
top_5_womens_products_sold
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
# Show the top 5 results.
womens_products_sold_pivot = pd.pivot_table(womens_products, values='units_sold', index=['retailer', 'region', 'state', 'city'],aggfunc='sum').reset_index()
womens_products_sold_pivot.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)
top_5_womens_products_sold_pivot = womens_products_sold_pivot.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)
top_5_womens_products_sold_pivot
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
# Optional: Rename the "total_sales" column to "Total Sales"
# Show the table.
sales_by_date = pd.pivot_table(combined_df, values='total_sales', index='invoice_date', aggfunc='sum').reset_index()
sales_by_date.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
sales_by_date
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
# Sort the resampled pivot table in descending order on "Total Sales".  
sales_by_date['invoice_date'] = pd.to_datetime(sales_by_date['invoice_date'])
sales_by_date.set_index('invoice_date', inplace=True)
daily_sales = sales_by_date.resample('D').sum().sort_values(by='Total Sales', ascending=False)
daily_sales
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
# Sort the resampled pivot table in descending order on "Total Sales".
weekly_sales = sales_by_date.resample('W').sum().sort_values(by='Total Sales', ascending=False)
weekly_sales
```
